# K-Means Clustering

## Importing the libraries

```python
import pandas as pdd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

Vamos excluir a coluna de costumer ID pois não faz sentido na nossa avaliação
No entanto, no sentido de ensino, vamos apenas ver as colunas de Annual Income e Spendingo Score para termos uma matrix bidimensional e ver graficamente os clusters.

```python
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:,[3,4]].values  #os parentises retos servem para indicar várias colunas, separadas por ,
print(X)
```

vamos criar uma variável dependente, que tomará um numero finito de valores, que será uma classe de valores que sta variável vai criar

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):  #quero testar 10 clusters
    kmeans = KMeans(n_clusters= k , init = 'k-means++', random_state = 42) #k-means++ para evitar a inicializaçáo random
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) #o atributo inerca_ dá o valor exato de wcss
plt.plot(range(1,11), wcss) #coordenadas de X,Y
plt.title("Elbow Method")
plt.xlabel('Number os clusters')
plt.ylabel("WCSS")
plt.show()
```

O numero ideal de clusters é 5.

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters= 5 , init = 'k-means++', random_state=42) #agora já sabemos o numero ideal de clusters e colocamos 5
y_kmeans = kmeans.fit_predict(X) #trina o modelo e retorna a variável dependente que iamos criar com os 5 clusters, cada grupo será grupado por similaridade
y_kmeans
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans== 0 ,0], X[y_kmeans== 0 , 1], s = 100, c= 'red', label = 'Cluster 1')  #X é a coluna de Annual Income e Y vai ser o Spending Score
plt.scatter(X[y_kmeans== 1 ,0], X[y_kmeans== 1 , 1], s = 100, c= 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans== 2 ,0], X[y_kmeans== 2 , 1], s = 100, c= 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans== 3 ,0], X[y_kmeans== 3 , 1], s = 100, c= 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans== 4 ,0], X[y_kmeans== 4 , 1], s = 100, c= 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, color = 'yellow', label = 'Centroids') #todas as linhas e a coluna 0 de X (Salary) e a coluna 1 (Y - Income)
plt.title("K-Means")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```

Rosa - low income and low spending

Cyan - high income and low spending

Green - low income and high spending

Blue - high income and high spending

Red - middle income and middle spending
